# Columbia Citibike Data Analysis

This notebook loads the filtered Citibike data for Columbia University area stations.

```python
import pandas as pd
import os
```

## Load Filtered Data

Load the pre-filtered dataset containing only trips involving Columbia University area stations.

```python
# Load the filtered data
data_path = os.path.join('..', 'data', 'columbia_filtered_citibike.csv')
df = pd.read_csv(data_path, parse_dates=['started_at', 'ended_at'])

print(f"Loaded DataFrame shape: {df.shape}")
print(f"Total rows: {df.shape[0]:,}")
print(f"Total columns: {df.shape[1]}")
```

## Example Visualization

Map of Columbia University area stations included in the filtered dataset.

```python
import plotly.express as px
```

```python
# Extract unique stations from start and end locations
start_stations = df[['start_station_id', 'start_station_name', 'start_lat', 'start_lng']].dropna()
start_stations.columns = ['station_id', 'station_name', 'lat', 'lng']

end_stations = df[['end_station_id', 'end_station_name', 'end_lat', 'end_lng']].dropna()
end_stations.columns = ['station_id', 'station_name', 'lat', 'lng']

# Combine and get unique stations
all_stations = pd.concat([start_stations, end_stations]).drop_duplicates(subset=['station_id'])

# Filter for Columbia stations only
columbia_station_ids = ["7783.18", "7741.04", "7745.07", "7727.07", "7713.11", "7692.11", "7713.01"]
columbia_stations = all_stations[all_stations['station_id'].isin(columbia_station_ids)]

print(f"Found {len(columbia_stations)} Columbia University area stations:")
print(columbia_stations[['station_id', 'station_name']].to_string(index=False))
```

```python
import plotly.express as px

fig = px.scatter_mapbox(
    columbia_stations,
    lat='lat',
    lon='lng',
    hover_name='station_name',
    hover_data={'station_id': True, 'lat': ':.6f', 'lng': ':.6f'},
    zoom=14,
    height=600,
    title='Columbia University Area Citi Bike Stations'
)

# Use OpenStreetMap tiles and center on Columbia
fig.update_layout(
    mapbox_style='open-street-map',
    mapbox_center={'lat': 40.807, 'lon': -73.962}
)

# Update marker appearance
fig.update_traces(marker=dict(size=12, color='blue'))

fig.show()
```

## Data Analysis

```python
# Drop missing station entries
df = df.dropna(subset=['start_station_id', 'end_station_id'])

# Calculate trip duration (in minutes)
df['trip_duration_min'] = (df['ended_at'] - df['started_at']).dt.total_seconds() / 60

# Extract time features
df['hour'] = df['started_at'].dt.hour
df['day_of_week'] = df['started_at'].dt.day_name()
df['month'] = df['started_at'].dt.month_name()
```

Monthly Exploration

```python
# Group by month
monthly_duration = (
    df.groupby('month')['trip_duration_min']
      .mean()
      .reset_index()
      .sort_values('month'))
```

```python
fig = px.bar(
    monthly_duration,
    x='month',
    y='trip_duration_min',
    title='Average Trip Duration per Month (Columbia Area)',
    labels={'trip_duration_min': 'Avg Duration (minutes)', 'month': 'Month'},
    text_auto='.1f',
    color='trip_duration_min',
    color_continuous_scale='Blues'
)

fig.update_layout(
    xaxis=dict(title='Month'),
    yaxis=dict(title='Average Trip Duration (min)'),
    template='plotly_white'
)

fig.show()
```

```python
monthly_user = (
    df.groupby(['month', 'member_casual'])['trip_duration_min']
      .mean()
      .reset_index()
)

fig = px.bar(
    monthly_user,
    x='month',
    y='trip_duration_min',
    color='member_casual',
    barmode='group',
    title='Average Trip Duration per Month by User Type'
)
fig.show()
```

Hourly Exploration

```python
rides_per_hour = df.groupby('hour').size().reset_index(name='trip_count')
px.bar(rides_per_hour, x='hour', y='trip_count', title='Rides per Hour of Day')
```

```python
rides_per_day = df.groupby('day_of_week').size().reset_index(name='trip_count')
px.bar(rides_per_day, x='day_of_week', y='trip_count', title='Rides per Day of Week')
```

```python
avg_duration_day = df.groupby('day_of_week')['trip_duration_min'].mean().reset_index()
px.bar(avg_duration_day, x='day_of_week', y='trip_duration_min', title='Avg Trip Duration by Day')
```

Most famous starting stations

```python
top_starts = (
    df.groupby('start_station_name').size().reset_index(name='trip_count')
      .sort_values('trip_count', ascending=False).head(10)
)
px.bar(top_starts, x='trip_count', y='start_station_name', orientation='h', title='Top 10 Start Stations')
```

Most popular ending stations

```python
top_end_stations = (
    df.groupby('end_station_name')
      .size()
      .reset_index(name='trip_count')
      .sort_values('trip_count', ascending=False)
      .head(10)
)

px.bar(top_end_stations, x='trip_count', y='end_station_name', orientation='h', title='Top 10 Ending Stations')
```

Most popular routes

```python
top_routes = (
    df.groupby(['start_station_name', 'end_station_name', 'start_lat', 'start_lng', 'end_lat', 'end_lng'])
      .size()
      .reset_index(name='trip_count')
      .sort_values('trip_count', ascending=False)
      .head(5)     
)
top_routes.head()
```

```python
import plotly.graph_objects as go

fig = go.Figure()

# Add each route as a line
for _, row in top_routes.iterrows():
    fig.add_trace(go.Scattermapbox(
        lat=[row['start_lat'], row['end_lat']],
        lon=[row['start_lng'], row['end_lng']],
        mode='lines+markers',
        line=dict(width=2, color='royalblue'),
        marker=dict(size=5),
        hoverinfo='text',
        text=f"{row['start_station_name']} → {row['end_station_name']}<br>Trips: {row['trip_count']}"
    ))

# Configure map layout
fig.update_layout(
    mapbox=dict(
        style='open-street-map',
        center=dict(lat=40.807, lon=-73.962),
        zoom=13
    ),
    title='Top Citi Bike Routes Around Columbia University',
    showlegend=False,
    margin=dict(l=0, r=0, t=40, b=0)
)

fig.show()
```
