# Citibike Data Loader

This notebook loads all Citibike trip data from the data/ folder into a single DataFrame.

```python
import pandas as pd
import glob
import os
```

## Load All Data Files

Find all CSV files in the data/ folder and load them into a single DataFrame.

```python
# Get all CSV files from data folder
data_path = os.path.join('..', 'data', '**', '*-citibike-tripdata_*.csv')
csv_files = glob.glob(data_path, recursive=True)

print(f"Found {len(csv_files)} CSV files")
print(f"\nFirst few files:")
for i, file in enumerate(csv_files[:5]):
	print(f"{i+1}. {os.path.basename(file)}")
```

```python
# Load all CSV files into a list
df_list = []

for i, file in enumerate(csv_files):
	print(f"Loading file {i+1}/{len(csv_files)}: {os.path.basename(file)}", end='\r')
	
	# Read CSV with datetime parsing
	df_temp = pd.read_csv(
		file,
		parse_dates=['started_at', 'ended_at']
	)
	
	df_list.append(df_temp)

print(f"\nLoaded {len(df_list)} files successfully")
```

```python
# Concatenate all DataFrames
df = pd.concat(df_list, ignore_index=True)

print(f"Combined DataFrame shape: {df.shape}")
print(f"Total rows: {df.shape[0]:,}")
print(f"Total columns: {df.shape[1]}")
```

## Filter for Columbia University Stations

Filter the dataset to include only trips that start or end at Columbia University area stations.

```python
columbia_stations = ["7783.18", "7741.04", "7745.07", "7727.07", "7713.11", "7692.11", "7713.01"]
filtered_df = df[(df['start_station_id'].isin(columbia_stations)) | (df['end_station_id'].isin(columbia_stations))]
```

### Sort and Clean Data

Sort the filtered data chronologically and reset the index.

```python
# Sort by started_at and ended_at
filtered_df_sorted = filtered_df.sort_values(by=['started_at', 'ended_at'])

print(f"Sorted DataFrame shape: {filtered_df_sorted.shape}")
filtered_df_sorted.head()
```

```python
# Reset index
filtered_df_sorted = filtered_df_sorted.reset_index(drop=True)

print(f"Index reset. New DataFrame shape: {filtered_df_sorted.shape}")
filtered_df_sorted.head()
```

### Save Filtered Data

Export the filtered and sorted dataset to a new CSV file.

```python
# Save to new CSV file
output_path = os.path.join('..', 'data', 'columbia_filtered_citibike.csv')
filtered_df_sorted.to_csv(output_path, index=False)

print(f"Saved {len(filtered_df_sorted):,} rows to: {output_path}")
print(f"File saved successfully!")
```
